```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
mbb=pd.read_csv("/content/MBB.csv")
mbb
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=mbb ['Price'], color='#c7675d', width=0.5, linewidth=2)
plt.title('Box Plot of MBB')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(mbb ['Price'],color='#c7675d', bins=20, kde=True)
plt.title('Histogram of MBB')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
