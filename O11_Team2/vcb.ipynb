```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
vcb=pd.read_csv("/content/VCB.csv",sep=";")
```

```python
vcb
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=vcb ['Price'], color='#c7675d', width=0.5, linewidth=2)
plt.title('Box Plot of VCB')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(vcb ['Price'],color='#c7675d', bins=20, kde=True)
plt.title('Histogram of VCB')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
