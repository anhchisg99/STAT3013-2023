```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
bidv=pd.read_csv("/content/BIDV.csv")
bidv
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=bidv ['Price'], color='#c7675d', width=0.5, linewidth=2)
plt.title('Box Plot of BIDV')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(bidv ['Price'],color='#c7675d', bins=20, kde=True)
plt.title('Histogram of BIDV')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
